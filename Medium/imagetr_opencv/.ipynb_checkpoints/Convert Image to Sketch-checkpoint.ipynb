### Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

### Read photo

```python
img=cv2.imread("mickey.jpg")
```

### Show image using openCV

```python
cv2.imshow('original image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Display using Matplotlib

```python
plt.imshow(img)
plt.show()
```

* The image is not consistant with the original image
* The reason is CV2 uses BGR color scheme whereas plt uses RGB scheme

### Convert BGR to RGB: Method 1

```python
plt.imshow(img[:,:,::-1])
plt.show()
```

### Convert BGR to RGB: Method 2

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

# Convert Image to a Pencil Sketch

```python
# Convert to Grey Image
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Invert Image
invert_img=cv2.bitwise_not(grey_img)
#invert_img=255-grey_img

# Blur image
blur_img=cv2.GaussianBlur(invert_img, (609,609),0)

# Invert Blurred Image
invblur_img=cv2.bitwise_not(blur_img)
#invblur_img=255-blur_img

# Sketch Image
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)

# Save Sketch 
cv2.imwrite('sketch.png', sketch_img)

# Display sketch
cv2.imshow('sketch image',sketch_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(14,8))

plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```

```python
def img2sketch(image_path):
    #Read image
    img=cv2.imread(image_path)
    
    # Convert to Grey Image
    grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Invert Image
    invert_img=cv2.bitwise_not(grey_img)
    #invert_img=255-grey_img

    # Blur image
    blur_img=cv2.GaussianBlur(invert_img, (609,609),0)

    # Invert Blurred Image
    invblur_img=cv2.bitwise_not(blur_img)
    #invblur_img=255-blur_img

    # Sketch Image
    sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)

    # Save Sketch 
    cv2.imwrite('sketch.png', sketch_img)

    # Display sketch
    cv2.imshow('sketch image',sketch_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
img=cv2.imread('kate_titanic.jpg')
cv2.imshow('original image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img2sketch('kate_titanic.jpg')
```
