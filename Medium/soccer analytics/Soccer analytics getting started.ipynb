```python
df=sb.events(match_id=7580)
lineup=sb.lineups(match_id=7580)
```

```python
lineup['Argentina'].player_nickname.isna()
```

```python
name_dict={}
teams=['France', 'Argentina']
for team in teams:
    for i in range(len(lineup[team])):
        key_n=lineup[team].player_name[i]
        if pd.isna(lineup[team].player_nickname[i])==True:
            val_n=lineup[team].player_name[i]
        else:
            val_n=lineup[team].player_nickname[i]
        name_dict.update({key_n:val_n})
```

```python
df_shots=df[df.shot_freeze_frame.isna()==False].reset_index()
```

```python
df_shots[df_shots.shot_outcome=='Goal']
```

```python
fig, ax= plt.subplots(figsize=(10,8))
pitch=Pitch(pitch_type='statsbomb', orientation='vertical', view='half',
              pitch_color='#22312b', line_color='#c7d5cc', figsize=(13, 12),
              constrained_layout=False, tight_layout=True)
pitch.draw(ax=ax)

shot_id=3
pitch.lines(xstart=df_shots.location[shot_id][0], 
            ystart=df_shots.location[shot_id][1],
            xend=df_shots.shot_end_location[shot_id][0],
            yend=df_shots.shot_end_location[shot_id][1],
            ax=ax, comet=True, cmap='jet')
if df_shots.shot_outcome[shot_id]=='Goal':
    pitch.scatter(df_shots.location[shot_id][0],df_shots.location[shot_id][1], marker='football', ax=ax)
else:
    pitch.scatter(df_shots.location[shot_id][0],df_shots.location[shot_id][1], color='green', ax=ax)
ff=pd.DataFrame(df_shots.shot_freeze_frame[shot_id])
for i in range(len(ff)):
    if ff.teammate[i]==True:
        pitch.scatter(x=ff.location[i][0],y=ff.location[i][1], ax=ax, color='green')
    if ff.teammate[i]==False:
        pitch.scatter(x=ff.location[i][0],y=ff.location[i][1], ax=ax, color='red')
plt.title(f'{df_shots.shot_outcome[shot_id]}, Time :{df_shots.minute[shot_id]}, Player : {name_dict[df_shots.player[shot_id]]}');
```
